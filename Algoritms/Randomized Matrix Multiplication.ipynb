```python
import numpy as np
import math
from time import time


#One pass random selection algorithm

def select(parray):
    d = 0
    ind = 0
    for i in range(len(parray)):
        d = d + parray[i]
        x = np.random.uniform(0,d)
        if x<parray[i]:
            ind = i
    return [ind,parray[ind]]
            

#Determine optimal probabilities

def probability_finder(A,B):
    probs=[]
    for i in range(len(A[0])):
        probs.append(np.linalg.norm(A[:,i])*np.linalg.norm(B[i]))
    probs = probs/sum(probs)
    return probs


A = np.random.rand(10000, 10000)#Might take a while to run with these numbers, can reduce them.
B = np.random.rand(10000, 10000)#If you reduce them, you should also reduce the numbers marked * below


def optimal_multiplication(A,B,c):
    p=probability_finder(A,B)
    selected = []
    scalars = []
    for i in range(c):
        x = select(p)
        selected.append(x[0])
        scalars.append(1/math.sqrt(c*x[1]))
    C = A[:, selected] * np.array(scalars)
    D = B[selected] * np.array(scalars)[:, np.newaxis]
    return np.matmul(C,D)

results = []
start_time = time()
AB=np.matmul(A,B)
time_taken_2 = time() - start_time
for i in range(10):#Test for various amounts selected
    print(100*(i+1))# * Reduce this if reducing matrix size
    start_time = time()
    CD=optimal_multiplication(A,B,100*(i+1))# * Reduce this if reducing matrix size
    time_taken_1 = time() - start_time
    print(time_taken_1)
    results.append([[time_taken_1,time_taken_2],np.linalg.norm(AB-CD)/np.linalg.norm(AB)])
results2 = results

for i in range(len(results)):#Get ratios of times
    results2[i][0]=results2[i][0][0]/results2[i][0][1]
    
for i in range(10):#Print results
    print("When selecting " + str(100*(i+1)) + " samples:")# * Reduce this if reducing matrix size
    print("Ratio of running time:" + str(results2[i][0]))
    print("Ratio of norms:" + str(results2[i][1]))
```
